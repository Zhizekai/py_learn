```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from factor_analyzer import FactorAnalyzer
data = load_iris()

%matplotlib inline
```

```python
X_train,X_test,y_train,y_test = train_test_split(data['data'],data['target'])
iris_dataframe = pd.DataFrame(X_train,columns=data.feature_names)
g = pd.plotting.scatter_matrix(iris_dataframe,
                               c=y_train,
                               figsize=(15,15),
                               marker='o',
                               hist_kwds={'bins':20},
                               s=60,
                               alpha=0.8
                              )
plt.savefig('./fff.png')
```

```python
df = pd.DataFrame(data=data.data,columns=data.feature_names)
```

```python
plt.style.use('ggplot')

x = data.data # 样本特征
y = data.target # 样本类型[0,1,2]
features = data.feature_names # 4个特征名称
targets = data.target_names

# df['petal length (cm)'][df['target']==0]
plt.figure(figsize=(10,4))
plt.plot(x[:,2][y==0],x[:,3][y==0] , 'bo',label=targets[0])  # bo是蓝色的O型圆圈
plt.plot(x[:,2][y==1],x[:,3][y==1] , 'kx',label=targets[1])
plt.plot(x[:,2][y==2],x[:,3][y==2] , 'ro',label=targets[2])

plt.xlabel(features[2])
plt.ylabel(features[3])
plt.title('Iris data set')
plt.legend()
plt.show()
```

```python
df = pd.DataFrame(data=data.data,columns=data.feature_names)
df['target']=data['target']
print('花的特征:',data.feature_names)
print('花的种类:',data.target_names)
```

```python
plt.style.use('ggplot')

x = data.data # 样本特征
y = data.target # 样本类型[0,1,2]
features = data.feature_names # 4个特征名称 长宽
targets = data.target_names # 花的种类

# df['petal length (cm)'][df['target']==0]
plt.figure(figsize=(10,4))

# #petal之间的关系
plt.plot(df['petal length (cm)'][df['target']==0],df['petal width (cm)'][df['target']==0], 'bo',label=targets[0])  # bo是蓝色的O型
plt.plot(df['petal length (cm)'][df['target']==1],df['petal width (cm)'][df['target']==1], 'kx',label=targets[1])
plt.plot(df['petal length (cm)'][df['target']==2],df['petal width (cm)'][df['target']==2], 'ro',label=targets[2])

plt.xlabel(features[2])
plt.ylabel(features[3])

plt.title('petal set')
plt.legend()
plt.show()

plt.figure(figsize=(10,4))
#sepal之间的关系
plt.plot(df['sepal length (cm)'][df['target']==0],df['sepal width (cm)'][df['target']==0],'bo',label=targets[0])
plt.plot(df['sepal length (cm)'][df['target']==1],df['sepal width (cm)'][df['target']==1],'ro',label=targets[1])
plt.plot(df['sepal length (cm)'][df['target']==2],df['sepal width (cm)'][df['target']==2],'k*',label=targets[2])
plt.xlabel(features[0])
plt.ylabel(features[1])

plt.title('sepal set')
plt.legend()
plt.show()
```
